```python
import numpy as np
import pandas as pd
import sklearn
from sklearn import linear_model


data = pd.read_csv("student-por.csv", sep=";")
data = data[["G1", "G2", "G3", "studytime", "failures", "absences"]]

predict = "G3"

x = np.array(data.drop([predict], axis=1))
y = np.array(data[predict])

x_train, x_test, y_train, y_test = sklearn.model_selection.train_test_split(x, y, test_size=0.1)

linear = linear_model.LinearRegression()

linear.fit(x_train, y_train)

accuracy = linear.score(x_test,y_test)
print("Model Accuracy: ", accuracy)

print("Coefficient: ", linear.coef_) #We have a 5 dimensional space here so we have 5 Coefficients! (our line starting position depending on 5 axis )
print("Intercept: ", linear.intercept_) # y

print("\n\n#####################################")
predictions = linear.predict(x_test)
for i in range(len(predictions)):
    print(f"Data #{i+1}")
    print("Model Prediction: ", predictions[i])
    print("Input Data: ", x_test[i])
    print("Real Label: ", y_test[i])
    print("#####################################")
    
    
#     Model Accuracy:  0.7338789388801772
# Coefficient:  [ 0.14080948  0.9834471  -0.23815192 -0.30036232  0.03282668]
# Intercept:  -1.2219170034987457


# #####################################
# Data #1
# Model Prediction:  18.744860429782598
# Input Data:  [19 18  2  0  2]
# Real Label:  18
# #####################################
# Data #2
# Model Prediction:  9.45968607430011
# Input Data:  [ 8 10  2  0  6]
# Real Label:  10
# #####################################
# Data #3
# Model Prediction:  4.7938663140511775
# Input Data:  [8 6 2 2 2]
# Real Label:  5
# #####################################
# Data #4
# Model Prediction:  12.581620337687895
# Input Data:  [14 13  4  0  0]
# Real Label:  14
# #####################################
# Data #5
# Model Prediction:  4.761596953765984
# Input Data:  [ 6  5  1  0 14]
# Real Label:  5
# #####################################
# Data #6
# Model Prediction:  11.724773396980472
# Input Data:  [13 11  3  1 40]
# Real Label:  11
# #####################################
# Data #7
# Model Prediction:  12.293623463488478
# Input Data:  [12 12  1  0  8]
# Real Label:  12
# #####################################
# Data #8
# Model Prediction:  9.29669027630225
# Input Data:  [ 9 10  3  0  4]
# Real Label:  10
# #####################################
# Data #9
# Model Prediction:  7.295831798084292
# Input Data:  [8 8 2 0 0]
# Real Label:  0
# #####################################
# Data #10
# Model Prediction:  18.973509584851417
# Input Data:  [18 18  1  0  6]
# Real Label:  18
# #####################################
```
